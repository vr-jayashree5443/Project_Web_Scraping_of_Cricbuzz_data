```python
url=["https://www.cricbuzz.com/cricket-stats/icc-rankings/men/batting", "https://www.cricbuzz.com/cricket-stats/icc-rankings/men/bowling", "https://www.cricbuzz.com/cricket-stats/icc-rankings/men/all-rounder"]
big_df=pd.DataFrame()
li=['batting', 'bowling','all-rounder' ]
for l,url in enumerate(url):
  req= requests.get(url)
  soup=BeautifulSoup(req.content)
  arr=[]
  for i in soup.find_all('div'):
    arr.append(i)
  # print(arr[115])
  arr_inner=[]
  for i in arr[115].find_all('div'):
    arr_inner.append(i)

  Position=[]
  Name=[]
  Country=[]
  Rating=[]
  count=1
  a=arr_inner
  i=4
  type="ODI"
  while(i<=len(a)-9):
    if count==1:
      i+=1
      count+=1
      # print("done")
    elif count==2:
      Position.append(re.sub('^<div.*">|</div>',"",str(a[i])))
      i+=1
      count+=1
    elif count==3:
      Name.append(re.sub('^<div.*Profile">|</a>.*',"",str(a[i])))
      Country.append(re.sub('^<div.*">|</div>',"",str(a[i])))
      i+=6
      count+=6
    elif count==9:
      Rating.append(re.sub('^<div.*">|</div>',"",str(a[i])))
      i+=1
      count=1
  x=li[l]
  df=pd.DataFrame({'Type':["ODI"]*len(Position[:92]),'Engaging':[x]*len(Position[:92]),'Position':Position[:92], 'Name':Name[:92], 'Country':Country[:92], 'Rating':Rating[:92]})
  big_df=pd.concat([big_df,df])

```
